```python
import pandas as pd
```

```python
raw_data = pd.read_csv('continuous dataset.csv')
raw_data
```

```python
data = raw_data.copy()
data['datetime'] = pd.to_datetime(data['datetime'])
data.set_index('datetime', inplace=True)
data
```

```python
data['Year'] = data.index.year
data['Month'] = data.index.month
data['Day'] = data.index.day
data['Hour'] = data.index.hour
data.columns.values
```

```python
data = data[['Year','Month', 'Day', 'Hour','nat_demand', 'T2M_toc', 'QV2M_toc', 'TQL_toc', 'W2M_toc',
       'T2M_san', 'QV2M_san', 'TQL_san', 'W2M_san', 'T2M_dav', 'QV2M_dav',
       'TQL_dav', 'W2M_dav', 'Holiday_ID', 'holiday', 'school']]
data
```

```python
from sklearn.preprocessing import StandardScaler
column_to_scaled = ['nat_demand', 'T2M_toc', 'QV2M_toc', 'TQL_toc', 'W2M_toc',
       'T2M_san', 'QV2M_san', 'TQL_san', 'W2M_san', 'T2M_dav', 'QV2M_dav',
       'TQL_dav', 'W2M_dav']

scaler = StandardScaler()
scaled = scaler.fit_transform(data[column_to_scaled])
scaled = pd.DataFrame(scaled, columns=column_to_scaled).reset_index(drop=True)
data_sceled = pd.concat([data[['Year','Month', 'Day', 'Hour','Holiday_ID', 'holiday', 'school']].reset_index(drop=True), scaled], axis= 1)
data_sceled
```

```python
data_input = data_sceled[['Year','Month', 'Day', 'Hour', 'T2M_toc', 'QV2M_toc', 'TQL_toc', 'W2M_toc',
       'T2M_san', 'QV2M_san', 'TQL_san', 'W2M_san', 'T2M_dav', 'QV2M_dav',
       'TQL_dav', 'W2M_dav', 'Holiday_ID', 'holiday', 'school']]
data_target = data_sceled['nat_demand']
data_target = pd.DataFrame(data_target, columns=['nat_demand'])
data_target
```

```python
n = len(data)
train_size = int(0.8*n)
validation_size = int(0.1*n)
test_size = n - train_size - validation_size

train_input = data_input[:train_size]
validation_input = data_input[train_size:train_size+validation_size]
test_input = data_input[train_size+validation_size:]

train_target = data_target[:train_size]
validation_target = data_target[train_size:train_size+validation_size]
test_target = data_target[train_size+validation_size:]
```

```python
import numpy as np

np.savez('Train_data.npz', inputs=train_input, targets=train_target)
np.savez('Validation_data.npz', inputs=validation_input, targets=validation_target)
np.savez('Test_data.npz', inputs=test_input, targets=test_target)
```
